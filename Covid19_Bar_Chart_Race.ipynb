```python
import pandas as pd
import numpy as np
import bar_chart_race as bcr
import datetime
```

```python
#Read from corona virus data csv
covid_data = pd.read_csv('opendata_eu_coronavirus.csv')
```

```python
#Clean through data
covid_data.head()
```

```python
print(list(covid_data.columns))
```

```python
covid_data.drop(columns=['geoId','countryterritoryCode','popData2019'], inplace=True)
```

### Covid Cases

```python
covid_cases = covid_data[['dateRep','cases','countriesAndTerritories']]
```

```python
#Convert all strings to datetime format
date = [datetime.datetime.strptime(i, '%d/%m/%Y') for i in covid_cases['dateRep']]
covid_cases['dateRep'] = date
```

```python
covid_cases
```

```python
covid_cases_df = covid_cases.pivot(index='dateRep', columns='countriesAndTerritories', values='cases')
```

```python
covid_cases_df.fillna(0)
```

```python
bcr.bar_chart_race(
        df=covid_cases_df, 
        filename='../docs/images/covid19_horiz.gif', 
        orientation='h', 
        sort='desc', 
        n_bars=8, 
        fixed_order=False, 
        fixed_max=True, 
        steps_per_period=20, 
        period_length=500, 
        end_period_pause=0,
        interpolate_period=False, 
        period_label={'x': .98, 'y': .3, 'ha': 'right', 'va': 'center'}, 
        period_template='%B %d, %Y', 
        period_summary_func=lambda v, r: {'x': .98, 'y': .2, 
                                          's': f'Total deaths: {v.sum():,.0f}', 
                                          'ha': 'right', 'size': 11}, 
        perpendicular_bar_func='median', 
        colors='dark12', 
        title='COVID-19 Deaths by Country', 
        bar_size=.95, 
        bar_textposition='inside',
        bar_texttemplate='{x:,.0f}', 
        bar_label_font=7, 
        tick_label_font=7, 
        tick_template='{x:,.0f}',
        shared_fontdict=None, 
        scale='linear', 
        fig=None, 
        writer=None, 
        bar_kwargs={'alpha': .7},
        fig_kwargs={'figsize': (6, 3.5), 'dpi': 144},
        filter_column_colors=False) 
```
